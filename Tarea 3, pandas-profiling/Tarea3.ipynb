#grafico


```
(ggplot(heart)
 + aes('age','trestbps')
 + aes(color='trestbps')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
 + ggtitle('Edad por presion(mm/Hg) segun sexo')
)
#resting blood pressure (in mm Hg on admission to the hospital)(trestbps)
```

```
(ggplot(heart)
 + aes('age','chol')
 + aes(color='chol')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
 + ggtitle('Edad por colesterol segun sexo')
)
#serum cholestoral in mg/dl (chol)
```

```
(ggplot(heart)
 + aes('age','thalach')
 + aes(color='thalach')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
 + ggtitle('Edad por el maximo archivado de latidos del corazon segun sexo')
)
#maximum heart rate achieved (thalach)
```

```
(ggplot(heart)
 + aes('age','oldpeak')
 + aes(color='oldpeak')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
)     
```

```
(ggplot(heart)
 + aes('age','ca')
 + aes(color='ca')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
 + ggtitle('Edad por ca segun sexo')
)
```

#Eliminar zeros de oldpeak



```
heart = heart.drop(heart[heart['oldpeak']==0].index)
```

```
(ggplot(heart)
 + aes('age','oldpeak')
 + aes(color='oldpeak')
 + geom_point()
 + stat_smooth()
 + facet_wrap('~sex')
)  
```

```
(ggplot(heart)
 + ggtitle('Number of major vessels (0-3) colored by flourosopy')
 + geom_bar(aes(x="ca"))
 + facet_wrap('sex')
)
```

```
(ggplot(heart)
 + aes('ca','age')
 + aes(color='ca', fill= 'age')
 + geom_bar(stat = 'identity')
# + stat_smooth()
 #+ labs(x='Edad', y='CA')
# + facet_grid(facets = '~ sexo')
)
```

```
(ggplot(heart)
 + geom_bar(aes(x='ca'))
 + aes (fill='age')
 +ggtitle('asdasdasd')
)
```

```
mujeres = heart[heart.sex == 0].groupby('ca').count()['age'].reset_index()
hombres = heart[heart.sex == 1].groupby('ca').count()['age'].reset_index()
df =pd.concat([mujeres,hombres],sort = 'False', keys = ['mujer', 'hombre'], names = ['sexo','id'])
df.loc['mujer'].plot(x = 'ca', y = 'age', kind = 'bar')
df.loc['hombre'].plot(x = 'ca', y = 'age', kind = 'bar')
```
